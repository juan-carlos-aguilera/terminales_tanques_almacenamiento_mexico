IMPORTACIÓN DE LIBRERÍAS

```python
# Librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Gráficos en línea
%matplotlib inline

# Formato de número
pd.set_option("display.float_format", "{:,.2f}".format)

# Estilo de fondo
sns.set_style(style='whitegrid')

# Estilo de escala y contexto
sns.set_context(context='notebook')

# Tamaño predeterminado a figuras
plt.rcParams['figure.figsize'] =(11,9.4)
```

IMPORTACIÓN DEL ARCHIVO

```python
tanques_terminales_df = pd.read_csv('../data/processed/tanques_terminales_clean.csv')
tanques_terminales_df.head()
```

ANÁLISIS BIVARIADO

Capacidad de Tanques (barriles) <= 100,000 barriles

```python
tanques_terminales_df[tanques_terminales_df['Capacidad_Tanque_barriles'] <= 100_000 ]
```

```python
sns.countplot(
    data=tanques_terminales_df[
        tanques_terminales_df['Capacidad_Tanque_barriles'] <= 100_000],
    x='Tipo_Tanque',
    palette='rocket'
)

for p in plt.gca().patches:
    plt.text(
        p.get_x() + p.get_width() / 2,
        p.get_height(),
        int(p.get_height()),
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

plt.title('Número y Tipos de Tanques vs Capacidad de Tanques <= 100,000 barriles',
          fontsize=14, fontweight='bold')

plt.xlabel('Tipo Tanque')
plt.ylabel('Número de Tanques')

plt.xticks(rotation=45, ha='right')
```

Capacidad de Tanques (barriles) > 100,000 barriles

```python
tanques_terminales_df[tanques_terminales_df['Capacidad_Tanque_barriles'] > 100_000]
```

```python
sns.countplot(
    data=tanques_terminales_df[
        tanques_terminales_df['Capacidad_Tanque_barriles'] > 100_000],
    x='Tipo_Tanque',
    palette='rocket'
)

for p in plt.gca().patches:
    plt.text(
        p.get_x() + p.get_width() / 2,
        p.get_height(),
        int(p.get_height()),
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

plt.title('Número y Tipos de Tanques vs Capacidad de Tanques > 100,000 barriles',
          fontsize=14, fontweight='bold')

plt.xlabel('Tipo Tanque')
plt.ylabel('Número de Tanques')

plt.xticks(rotation=45, ha='right')
```

Cantidad de Subtipo de Tanque vs Tipos de Tanques

```python
sns.countplot(
    data=tanques_terminales_df,
    x='Tipo_Tanque',
    hue='Subtipo_Tanque',
    palette='rocket',
)

for p in plt.gca().patches:
    plt.text(
        p.get_x() + p.get_width() / 2,
        p.get_height(),
        int(p.get_height()),
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

plt.title('Subtipo de Tanques vs Tipos de Tanques',
          fontsize=14, fontweight='bold')

plt.xlabel('Tipo Tanque')
plt.ylabel('Número de Tanques')

plt.xticks(rotation=45, ha='right')

plt.savefig('../reports/figures/03_tanques/03_tipo_subtipo_tanques.png',
            dpi=300, bbox_inches='tight')
```

RESUMEN DE ANÁLISIS BIVARIABLE

Capacidad del Tanque (Barriles) por Tipo de Tanque

```python
grouped_table = (
    tanques_terminales_df
    .groupby(['Tipo_Tanque'], as_index=False)
    .agg({
        'Capacidad_Tanque_barriles': 'sum'
    })
)

grouped_table['Proporción_Capacidad_Tanque_(%)'] = (
    grouped_table['Capacidad_Tanque_barriles'] / (grouped_table['Capacidad_Tanque_barriles'].sum()) * 100
)

grouped_table.sort_values(
    by='Capacidad_Tanque_barriles', ascending=False
)
```

Capacidad del Tanque (barriles) por Subtipo de Tanque

```python
grouped_table = (
    tanques_terminales_df
    .groupby(['Subtipo_Tanque'], as_index=False)
    .agg({
        'Capacidad_Tanque_barriles': 'sum'
    })
)

grouped_table['Proporción_Capacidad_Tanque_(%)'] = (
    grouped_table['Capacidad_Tanque_barriles'] / (grouped_table['Capacidad_Tanque_barriles'].sum()) * 100
)

grouped_table.sort_values(
    by='Capacidad_Tanque_barriles', ascending=False
)
```
