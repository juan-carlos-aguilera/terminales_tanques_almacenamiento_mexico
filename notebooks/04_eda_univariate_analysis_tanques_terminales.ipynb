RESÚMEN DE ANÁLISIS UNIVARIABLE

Variables Numéricas - Capacidad de Tanques (barriles)

```python
numeric_variables = ['Capacidad_Tanque_barriles']

summary_table = []

for var in numeric_variables:
    total_len= len(tanques_terminales_df)
    minor_equal = (tanques_terminales_df[var] <= 100_000).sum()
    mayor = (tanques_terminales_df[var] > 100_000).sum()

    summary_table.append({
        'variable': var,
        '> 100,000 barriles (N)': mayor,
        '<= 100,000 barriles(N)': minor_equal,
        '> 100,000 barriles (%)': round((mayor/total_len) * 100,2),
        '<= 100,000 barriles (%)': round((minor_equal/total_len) * 100,2),
        'Promedio (barriles)': tanques_terminales_df[var].mean(),
        'Mediana (barriles)': tanques_terminales_df[var].median(),
        'Desviación Estándar': tanques_terminales_df[var].std()
    })

summary_numeric_table = pd.DataFrame(summary_table)
summary_numeric_table
```

Variables Categóricas (Tipo Tanque y Subtipo Tanque)

```python
category_variables = ['Tipo_Tanque', 'Subtipo_Tanque']

summary_category = {}

for var in category_variables:
    table = (
        tanques_terminales_df[var]
        .value_counts()
        .rename_axis(var)
        .reset_index(name = 'Cantidad_Tanques')
    )

    table['Proporcion (%)'] = (
        table['Cantidad_Tanques']/table['Cantidad_Tanques'].sum() * 100
    )

    summary_category[var] = table

for var, table in summary_category.items():
    print(f'Distribución de Tanques por: {var}')
    display(table)
```
