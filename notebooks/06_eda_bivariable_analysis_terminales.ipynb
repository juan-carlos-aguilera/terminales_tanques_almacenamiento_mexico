RESUMEN DE ANÁLISIS BIVARIABLE

Región

```python
grouped_table = (
    terminales_mexico_df
    .groupby(['Región'], as_index=False)
    .agg({
        'Capacidad_Diseño_barriles': 'sum',
        'Capacidad_Operativa_barriles':  'sum',
        'Nombre_Terminal': 'count'
    })
    .rename(columns={'Nombre_Terminal': 'N°_Terminales'})
)

total_diseño = grouped_table['Capacidad_Diseño_barriles'].sum()
total_operativa = grouped_table['Capacidad_Operativa_barriles'].sum()


grouped_table['%_Capacidad_Diseño'] = (
    grouped_table['Capacidad_Diseño_barriles'] / total_diseño * 100
)

grouped_table['%_Capacidad_Operativa'] = (
    grouped_table['Capacidad_Operativa_barriles'] / total_operativa * 100
)


grouped_table = grouped_table.sort_values(by='Capacidad_Diseño_barriles', ascending=False)

grouped_table
```

Tipo de Terminal

```python
grouped_table = (
    terminales_mexico_df
    .groupby(['Tipo_Terminal'], as_index=False)
    .agg({
        'Capacidad_Diseño_barriles': 'sum',
        'Capacidad_Operativa_barriles':  'sum',
        'Nombre_Terminal': 'count'
    })
    .rename(columns={'Nombre_Terminal': 'N°_Terminales'})
)

total_diseño = grouped_table['Capacidad_Diseño_barriles'].sum()
total_operativa = grouped_table['Capacidad_Operativa_barriles'].sum()


grouped_table['%_Capacidad_Diseño'] = (
    grouped_table['Capacidad_Diseño_barriles'] / total_diseño * 100
)

grouped_table['%_Capacidad_Operativa'] = (
    grouped_table['Capacidad_Operativa_barriles'] / total_operativa * 100
)


grouped_table = grouped_table.sort_values(by='Capacidad_Diseño_barriles', ascending=False)

grouped_table
```

Tipos de Productos en Almacenar

```python
grouped_table = (
    terminales_mexico_df
    .groupby(['Tipos_Productos_Almacenar'], as_index=False)
    .agg({
        'Capacidad_Diseño_barriles': 'sum',
        'Capacidad_Operativa_barriles':  'sum',
        'Nombre_Terminal': 'count'
    })
    .rename(columns={'Nombre_Terminal': 'N°_Terminales'})
)

total_diseño = grouped_table['Capacidad_Diseño_barriles'].sum()
total_operativa = grouped_table['Capacidad_Operativa_barriles'].sum()


grouped_table['%_Capacidad_Diseño'] = (
    grouped_table['Capacidad_Diseño_barriles'] / total_diseño * 100
)

grouped_table['%_Capacidad_Operativa'] = (
    grouped_table['Capacidad_Operativa_barriles'] / total_operativa * 100
)


grouped_table = grouped_table.sort_values(by='Capacidad_Diseño_barriles', ascending=False)

grouped_table
```

Porcentaje de Aprovechamiento Operativo (%) - Terminal

```python
terminales_mexico_df['Aprovechamiento_Operativo_(%)'] = (
    terminales_mexico_df['Capacidad_Operativa_barriles']/terminales_mexico_df['Capacidad_Diseño_barriles'] * 100
)

terminales_mexico_df[
    ['Nombre_Terminal', 'Capacidad_Diseño_barriles','Capacidad_Operativa_barriles', 'Aprovechamiento_Operativo_(%)']
].sort_values(by='Aprovechamiento_Operativo_(%)', ascending=False)
```

```python
sns.scatterplot(
    data=terminales_mexico_df,
    x='Capacidad_Diseño_barriles',
    y='Capacidad_Operativa_barriles',
    alpha = 0.6
)

plt.plot(
    [0, terminales_mexico_df['Capacidad_Diseño_barriles'].max()],
    [0, terminales_mexico_df['Capacidad_Operativa_barriles'].max()],
    color='red', linestyle='dashed', label='Línea de igualdad y = x'
)
plt.title('Relación entre Capacidad de Diseño y Capacidad Operativa',
            fontsize=10, fontweight='bold')

plt.xlabel('Capacidad de Diseño (barriles)')
plt.ylabel('Capacidad Operativa (barriles)')
plt.legend()

plt.savefig('../reports/figures/04_relaciones/01_correlacion_capacidades.png',
            dpi=300, bbox_inches='tight')
```

```python
aprovechamiento_region = (
    terminales_mexico_df
    .groupby('Región', as_index= False)
    .agg({
        'Capacidad_Diseño_barriles': 'sum',
        'Capacidad_Operativa_barriles': 'sum'
    })
)

aprovechamiento_region['Aprovechamiento_Operativo_%'] = (
    aprovechamiento_region['Capacidad_Operativa_barriles'] / aprovechamiento_region['Capacidad_Diseño_barriles'] * 100
)

aprovechamiento_region.sort_values(['Aprovechamiento_Operativo_%'], ascending= False)
```
